# Tutorial para la biblioteca PandaSQL

```python
import pandas as pd
import matplotlib.pyplot as plt
#pd.set_printoptions(max_columns=25)
from pandasql import sqldf
from pandasql import load_meat, load_births
from pandasql import sqldf
```

## Construccion de Data-sets

Usando dos ejemplos de bases de datos que vienen en la libreria 'pandasql', se crearan dos bases de datos:

meat: Conjunto de datos del departamento de agricultura de EE.UU. que contiene mediciones sobre perspectivas y producción de ganado, productos lácteos y aves de corral.

births: conjunto de datos de la division de estadistica de las Naciones Unidas que contiene estadísticas demográficas de nacimientos vivos por mes.

```python
meat = load_meat()
births = load_births()
```

```python
print (meat.head())
```

```python
print (births.head())
```

Ejemplo de filrado de datos desde las bases de datos usando el lenguaje de SQL:

```python
pysqldf = lambda q: sqldf(q, globals())
q = """
    SELECT
        m.date
        , m.beef
        , b.births
    FROM
        meat m
    LEFT JOIN
        births b
            ON m.date = b.date
    WHERE
        m.date > '1974-12-31';
"""
```

Impresion de los datos filtrados en una grfica que muestra una relacion entre el ganado que se sacrifica y el numero de nacimientos en funcion del tiempo. 

```python
df = pysqldf(q)
df.births = df.births.fillna(method='backfill')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(df['beef'].rolling(12).mean(), color='b')
ax1.set_xlabel('Meses a partir de 1975')
ax1.set_ylabel('Ganado sacrificado', color='b')
ax2 = ax1.twinx()
ax2.plot(df['births'].rolling(12).mean(), color='r')
ax2.set_ylabel('Bebes nacidos', color='r')
plt.title("Consumo de carne y Tasa de nacimiento")
plt.show()
```

## Uso basico

Escribir SQL y ejecutarlo en el Dataframe de pandas sustituyendo Dataframes por tablas.

```python
q = """
SELECT
*
FROM
meat
LIMIT 10;"""
print (sqldf(q, locals()))
```

### Agregacion

pandasql admite agregacion. Se pueden usar nombres de columnas con alias o numeros de columna en un group by. Por ejemplo:

```python
q = """ 

    SELECT
        strftime("%Y", date)
        , SUM(births)
    FROM births
    GROUP BY 1
    ORDER BY 1; 
"""

print (sqldf(q, locals()))
```

### locals() vs. globals()

pandasql necesita tener acceso a otras variables en su sesion / entorno. Se pueden pasar variables locales, locals(), a pandasql al ejecutar una instrucción SQL, pero si se estan ejecutando muchas consultas, podria ser un problema. Para evitar pasar a los locales todo el tiempo, se puede agregar esta función auxiliar al script para establecer variables globales, globals(), de esta manera:

```python
def pysqldf(q):
    return sqldf(q, globals())
   
q = """
    SELECT
        *
    FROM
        births
    LIMIT 10;"""
print (pysqldf(q))
```

### joins

Ejemplo de como unir dataframes usando la sintaxis de SQL:

```python
q = """
    SELECT
        m.date
        , b.births
        , m.beef
    FROM
        meat m
    INNER JOIN
        births b
            on m.date = b.date
    ORDER BY
        m.date
    LIMIT 100;
    """

joined = pysqldf(q)
print (joined.head())
```

### Where

Ejemplo del uso de la funcion 'where' para hacer un filtro de una base de datos:

```python
q = """
    SELECT
        date
        , beef
        , veal
        , pork
        , lamb_and_mutton
    FROM
        meat
    WHERE
        lamb_and_mutton >= veal
    ORDER BY date DESC
    LIMIT 10;
    """
print (pysqldf(q))
```

## Ejemplos de scripts comunmente usados en SQL como consultas, ordenamiento, funciones y uniones:

#### - Ordenar de forma 'aleatoria'

```python
q = """SELECT 
    *
    FROM
        meat
    ORDER BY RANDOM()
    LIMIT 10;"""
print (pysqldf(q))
```

#### - Unir todo

```python
q = """
        SELECT
            date
            , 'beef' AS meat_type
            , beef AS value
        FROM meat
        UNION ALL
        SELECT
            date
            , 'veal' AS meat_type
            , veal AS value
        FROM meat
        UNION ALL
        SELECT
            date
            , 'pork' AS meat_type
            , pork AS value
        FROM meat
        UNION ALL
        SELECT
            date
            , 'lamb_and_mutton' AS meat_type
            , lamb_and_mutton AS value
        FROM meat
        ORDER BY 1
    """
print (pysqldf(q).head(20))
```

#### - Consultas

```python
q = """
    SELECT 
        m1.date
        , m1.beef 
    FROM 
        meat m1 
    WHERE m1.date IN 
        (SELECT 
            date 
        FROM meat 
        WHERE 
            beef >= broilers 
        ORDER BY date)
"""
more_beef_than_broilers = pysqldf(q)
print (more_beef_than_broilers.head(10))
```
