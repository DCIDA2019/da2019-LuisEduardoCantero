# Busca dos paginas en internet donde la base de datos no esté disponible para descargarse sino que este en forma de tabla. Descargalas como en el ejemplo de clase.

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import requests
import os
```

### Pagina de horarios de licenciatura en la DCI

```python
file = 'HORARIOS.csv'
url = 'http://www.dci.ugto.mx/estudiantes/index.php/mcursos/horarios-licenciatura'
if not os.path.isfile(file):  
    html = requests.get(url).content
    df_list = pd.read_html(html,header=1)
    lista = df_list[-1]
    lista.to_csv('HORARIOS.csv')
else:
    lista = pd.read_csv(file)
```
