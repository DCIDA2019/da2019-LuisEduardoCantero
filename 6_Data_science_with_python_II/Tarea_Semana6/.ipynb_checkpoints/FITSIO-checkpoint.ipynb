## Graficamos los datos

```python
ra = data1_tab['ra']
dec = data1_tab['dec']
```

```python
data1 = hdul[1].data
```

```python
plt.plot(data1['ra'][::1], data1['dec'],'.', c='g')
plt.rcParams["figure.figsize"] = (12, 7)
```

```python
colors = ['black','blue', 'darkcyan', 'green', 'yellow']
```

```python
plt.rcParams["figure.figsize"] = (16, 7)
plt.gca().invert_xaxis()
plt.scatter(data1['ra'], data1['dec'], s=1, label = None, c = data1['nexp_g'], cmap = ListedColormap(colors))
plt.axis(aspect='equal')
plt.xlabel('Ra (deg)')
plt.ylabel('Dec (deg)')
plt.colorbar(label = 'Number of exposures')
plt.title('MzLS + BASS DR4: Number of exposures in g')
plt.clim(0, 5.5)
```

```python
r = np.delete(data1['ra'], np.where((data1['nexp_g'] < 1 )))
d = np.delete(data1['dec'], np.where((data1['nexp_g'] < 1) ))
nexp_g = np.delete(data1['nexp_g'], np.where((data1['nexp_g'] < 1)))
```

```python
len(r)
```

```python
len(d)
```

```python
plt.rcParams["figure.figsize"] = (12, 7)
plt.gca().invert_xaxis()
plt.scatter(r, d, s=1, label = None, c = nexp_g, cmap = ListedColormap(colors))
plt.axis(aspect='equal')
plt.xlabel('ra (deg)')
plt.ylabel('dec (deg)')
plt.colorbar(label = 'Number of exposures')
plt.title('MzLS + BASS DR4: Number of exposures in g')
plt.clim(0, 5.5)
```

```python
plt.rcParams["figure.figsize"] = (13, 6.5)
plt.scatter(r, d, s=1, label = None, c = nexp_g, cmap = plt.cm.get_cmap('viridis', 5))
plt.axis(aspect='equal')
plt.xlabel('Ra (deg)')
plt.ylabel('Dec (deg)')
plt.colorbar(label = 'Number of exposures')
plt.title('MzLS + BASS DR4: Number of exposures in g')
plt.clim(0, 5.6)
plt.xlim(90, 310)
plt.ylim(30, 80)
plt.gca().invert_xaxis()
```
