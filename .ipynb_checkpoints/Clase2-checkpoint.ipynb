```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import requests
import os
```

```python
x1=np.arange(1, 50, 0.3)

def f1(x):
    x2 = []
    for i in range(0, len(x)):
        if  x[i]%2 ==0 :
            x2.append(x[i]*x[i])
        if  x[i]%2 !=0 :
            x2.append(np.log(x[i]))
    return x2
```

```python
y1 = f1(x1)
```

```python
plt.rcParams["figure.figsize"] = (8,8)
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.plot(x1, y1, "-", color = 'k')
plt.grid(b=None, which='major', axis='both')
plt.legend()
```

```python
def g(x):
    if x > 0:
        return np.cos(x)
    else:
        return np.sin(x)
g1 = np.vectorize(g)    
```

```python
x2 = np.linspace(-2*np.pi, 2*np.pi, 50)
y2 = g1(x2)
```

```python
plt.rcParams["figure.figsize"] = (8,8)
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.plot(x2, y2, "-", color = 'k')
plt.grid(b=None, which='major', axis='both')
plt.legend()
```

```python
y = np.zeros_like(x2)
w = x2 > 0
%timeit y[w] = g1(x2[w])
%timeit y[~w] = g1(x2[~w])
```

```python
y2 = np.asarray([g(x) for x in x2])
%timeit y2
```

```python
def g2(x):
    y = np.zeros_like(x)
    for i in range(len(x)):
        if x[i]>0:
            y[i] = np.cos(x[i])
        if x[i]>0:
            y[i] = np.sin(x[i])
    return y
            
```
