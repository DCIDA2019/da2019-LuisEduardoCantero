## - ¿Cuantos días ha llovido?

```python
q = """SELECT
            COUNT(precipi)
        FROM
            CLIMA
        WHERE
            precipi > 0;"""
print (sqldf(q, locals()))
```

## - ¿Cuál ha sido la temperatura máxima para días con neblina y dias sin neblina?

```python
q = """SELECT
            COUNT(fog)
        FROM
            CLIMA
        WHERE
            fog > 0;"""
print (sqldf(q, locals()))
```

```python
q = """SELECT
            COUNT(fog)
        FROM
            CLIMA
        WHERE
            fog = 0;"""
print (sqldf(q, locals()))
```
