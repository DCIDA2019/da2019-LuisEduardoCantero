# Reproduce las graficas de la distribucion uniforme de puntos dentro de un circulo y dentro de una esfera.

```python
# Aqui cargamos las liberias a usar

from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import scipy.special
import numpy as np
import math
%matplotlib inline
```

### Generamos 'x' y 'y' a partir de una distribucion uniforme de tal forma que $r=sqry(x^2+y^2)<=1$

```python
tam = 10000 #Tamano de la muestra
```

```python
fx = np.random.uniform(-1, 1, tam)
fy = np.random.uniform(-1, 1, tam)
x1 = np.empty(tam)
y1 = np.empty(tam)
n1 = np.empty(tam)
m1 = np.empty(tam)
```

- Graficamos los puntos aleatorios generados. El ciclo es para hacer distincion entre los puntos que estan dentro de la circunferencia buscada

```python
for j in range(0, tam):
    radio = math.sqrt(math.pow(fx[j], 2) + math.pow(fy[j], 2))
    if radio <= 1:
                 x1[j] = fx[j]
                 y1[j] = fy[j]
                 n1[j] = -1 # movemos estos puntos a la coordenada (-1,-1) para que no queden con un valor nulo.
                 m1[j] = -1
    else:
                 n1[j] = fx[j]
                 m1[j] = fy[j]
                 x1[j] = -1 # movemos estos puntos a la coordenada (-1,-1) para que no queden con un valor nulo.
                 y1[j] = -1
                 
```

```python
plt.rcParams["figure.figsize"] = (8,8)
plt.plot(x1,y1, "*")
plt.plot(n1,m1, "*")
plt.xlabel("EJE X")
plt.ylabel("EJE Y")
```

### Generamos $\sqrt{r}$ y theta a apartir de una distribucion uniforme

```python
r = np.random.uniform(0, 1, tam)
theta = np.random.uniform(0, 2 * 3.1415926535, tam)
xx = np.empty(tam)
yy = np.empty(tam)
xx1 = np.empty(tam)
yy1 = np.empty(tam)
nn = np.empty(tam)
mm = np.empty(tam)
for i in range(0, tam):
    xx[i] = math.sqrt(-2*math.log(r[i])) * math.cos(theta[i])
    yy[i] = math.sqrt(-2*math.log(r[i])) * math.sin(theta[i])
```

```python
for j in range(0, tam):
    rad = math.sqrt(math.pow(xx[j], 2) + math.pow(yy[j], 2))
    if rad <= 1:
                 xx1[j] = xx[j]
                 yy1[j] = yy[j]
                 nn[j] = -1 # movemos estos puntos a la coordenada (-1,-1) para que no queden con un valor nulo.
                 mm[j] = -1
    else:
                 nn[j] = xx[j]
                 mm[j] = yy[j]
                 xx1[j] = -1 # movemos estos puntos a la coordenada (-1,-1) para que no queden con un valor nulo.
                 yy1[j] = -1
                 
                 
```

```python
plt.rcParams["figure.figsize"] = (8,8)
plt.plot(xx1,yy1, "*")
plt.plot(nn,mm, "*")
plt.xlabel("EJE X")
plt.ylabel("EJE Y")
plt.xlim(-1,1)
plt.ylim(-1,1)
```
